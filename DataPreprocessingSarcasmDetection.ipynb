```python
import json
import nltk

from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
from nltk.stem import PorterStemmer
```

```python
def filter_stop_words(words):
    stop_words = set(stopwords.words('english'))
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return filtered_words
```

```python
def lemmatize_words(words):
    lemmatizer = WordNetLemmatizer()
    lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
    return lemmatized_words
```

```python
def stem_words(words):
    stemmer = PorterStemmer()
    stemmed_words = [stemmer.stem(word) for word in words]
    return stemmed_words
```

```python
path='/content/gdrive/MyDrive/Sarcasm_Headlines_Dataset.json'
fp=open(path, "r+")
line_data = fp.readlines()

raw_data_json = []
for it in line_data:
  raw_data_json.append(json.loads(it))

print("Raw data is :", raw_data_json[:100])
```

```python
X = []
Y = []
special_characters = "!@#$%^&*()-_+=[]{}|:;\"'<>,.?/"
for elem in raw_data_json:
  text = elem['headline']
  outcome = elem['is_sarcastic']

  x=[]
  str1 = ""
  for char in text:
    if char == ' ':
      x.append(str1)
      str1 = ""
    elif char not in special_characters:
      str1 = str1 + char

  if len(x) > 0:
    filtered_x = filter_stop_words(x)
  else:
    continue
  if len(filtered_x) > 0:
    stemed_x = stem_words(filtered_x)
  else:
    continue
  if len(stemed_x) > 0:
    lemmatized_x = lemmatize_words(stemed_x)
    X.append(lemmatized_x)
    Y.append(outcome)
```

```python
assert (len(X) == len(Y))
(X_train, Y_train)=(X[:20020], Y[:20020])
(X_test, Y_test)=(X[20020:], Y[20020:])

for i in range(500):
  print(X_train[i], Y_train[i])
```
